### Define the ground parameters for image, and model that will be build

```python
parameters={
    'image_size':(80,80), # untuk ukuran gambar yang mau dilatih
    'batch_size':32, # Banyaknya data yang mau dilatih dalam 1 epoch
    'epoch':3, # banyaknya iterasi training model
    'optimizer':'adam', # optimizer
    'loss_fn':'categorical_crossentropy', # loss function
    'metrics':['accuracy'] # metrics evalaution -> untuk menilai seberapa tepat model
}
```

### Create train dataset and look up for random sample

```python
train_data=Dataset('./Data/Train/',image_size=parameters['image_size'],batch_size=8)
train_data.getSample()
```

### Create Test dataset and look up for random sample

```python
test_data=Dataset('./Data/Test',image_size=parameters['image_size'],batch_size=8)
test_data.getSample()
```

```python
model=CustomModel(parameters=parameters)
```

```python
history=model.trainModel(train_data=train_data.data)
history
```

```python
model.evaluateModel(test_data=test_data.data)
```
